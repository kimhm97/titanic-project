```python
# ============================================================
# 1. 라이브러리 불러오기
# ============================================================

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

# 주피터 노트북 안에서 그래프가 바로 보이도록 설정한다. 
%matplotlib inline

# 그래프 기본 테마를 설정해서 시각적으로 보기 좋게 만든다. 
sns.set_theme(style="whitegrid")

# 출력되는 컬럼 수 제한을 풀어서 모든 컬럼을 한 번에 볼 수 있게 한다
pd.set_option('display.max_columns', None)

# 실수형 숫자를 소수점 둘째 자리까지 표시하도록 설정한다.
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# ============================================================
# 2. 데이터 불러오기
# ============================================================

train_path = "../../data/raw/train.csv"
test_path = "../../data/raw/test.csv"
```

```python
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
# ============================================================
# 3. 데이터 샘플 확인
# ============================================================

print("[train 상위 5행]")
display(train.head())

print("\n[test 상위 5행]")
display(test.head())
```

```python
# ============================================================
# 4. 데이터 구조 확인
# ============================================================

print("[train shape]")
print(train.shape)

print("\n[test shape]")
print(test.shape)

print("\n[train info]")
train.info()

print("\n[test info]")
test.info()
```

```python
# ============================================================
# 5. 기본 통계량 확인
# 숫자형 변수의 개수, 평균, 표준편차, 최소값, 사분위수, 최대값을 본다.
# ============================================================

print("[train 기본 통계량]")
display(train.describe())

print("\n[test 기본 통계량]")
display(test.describe())
```

```python
# ============================================================
# 6. 컬럼명 및 자료형 확인
# ============================================================

print("[train 컬럼명]")
print(train.columns.tolist())

print("\n[test 컬럼명]")
print(test.columns.tolist())

print("\n[train dtype]")
print(train.dtypes)

print("\n[test dtype]")
print(test.dtypes)
```

```python
# ============================================================
# 7. 숫자형 / 범주형 컬럼 분리
# ============================================================

train_num_cols = train.select_dtypes(include=['int64','float64']).columns.tolist()
train_obj_cols = train.select_dtypes(include=['object']).columns.tolist()

test_num_cols = test.select_dtypes(include=['int64','float64']).columns.tolist()
test_obj_cols = test.select_dtypes(include=['object']).columns.tolist()

print("[train 숫자형 컬럼]")
print(train_num_cols)

print("\n[train 범주형 컬럼]")
print(train_obj_cols)

print("\n[test 숫자형 컬럼]")
print(test_num_cols)

print("\n[test 범주형 컬럼]")
print(test_obj_cols)
```

```python
# ============================================================
# 8. 컬럼별 고유값 개수 확인
# 각 컬럼의 고유값 개수를 오름차순으로 정렬해서 확인한다.
# 어떤 컬럼이 식별자에 가깝고, 어떤 컬럼이 범주형인지 감을 잡을 수 있다
# ============================================================

print("[train 컬럼별 고유값 개수]")
print(train.nunique(dropna=False).sort_values())

print("\n[test 컬럼별 고유값 개수]")
print(test.nunique(dropna=False).sort_values())
```

```python
# ============================================================
# 9. 결측값 개수 확인
# ============================================================

print("[train 결측값 개수]")
print(train.isnull().sum().sort_values(ascending=False))

print("\n[test 결측값 개수]")
print(test.isnull().sum().sort_values(ascending=False))
```

```python
# ============================================================
# 10. 중복 행 확인
# ============================================================

print("[train 중복 행 개수]")
print(train.duplicated().sum())

print("\n[test 중복 행 개수]")
print(test.duplicated().sum())
```

```python
# ============================================================
# 11. train / test 컬럼 차이 확인
# ============================================================

train_only_cols = set(train.columns) - set(test.columns)
test_only_cols = set(test.columns) - set(train.columns)

print("[train에만 있는 컬럼]")
print(train_only_cols)

print("\n[test에만 있는 컬럼]")
print(test_only_cols)
```

```python
# ============================================================
# 12. 기본 요약 테이블 생성
# ============================================================

basic_info_summary = pd.DataFrame({
    'train_dtype' : train.dtypes.astype(str),
    'train_nunique' : train.nunique(dropna=False),
    'train_missing_count' : train.isnull().sum(),
    'train_missing_ratio':(train.isnull().sum() / len(train)) * 100
})

print("[basic_info_summary]")
display(basic_info_summary)

basic_info_summary.to_csv("../../output/tables/basic_info_summary.csv", index=True)

print("저장완료")
```
