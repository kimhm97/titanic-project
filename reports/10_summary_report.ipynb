```python
# ============================================================
# 1. 라이브러리 불러오기 
# ============================================================

import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
# ============================================================
# 2. 저장된 결과 파일 불러오기 
# ============================================================

univariate_summary = pd.read_csv("../output/tables/univariate_summary.csv", index_col=0)

cat_test = pd.read_csv("../output/tables/categorical_statistical_tests.csv")
num_test = pd.read_csv("../output/tables/numerical_statistical_tests.csv")

feature_set_summary = pd.read_csv("../output/tables/feature_set_cv_summary.csv")
model_comparison = pd.read_csv("../output/tables/model_comparison_results.csv")

submission = pd.read_csv("../output/submissions/titanic_submission.csv")
```

```python
# ============================================================
# 3. 전처리 후 단변량 분석 요약
# ============================================================

print("=" * 60)
print("전처리 후 단변량 분석 요약")
print("=" * 60)

print("\n[전처리 후 결측값 확인]")
display(univariate_summary[['missing_count','missing_ratio']])

print("\n[결측값이 남아있는 컬럼]")
display(univariate_summary[univariate_summary['missing_count'] > 0])

print("\n[왜도 높은 컬럼 Top]")
display(univariate_summary
        .assign(abs_skew = univariate_summary['skew'].abs())
        .sort_values(by='abs_skew', ascending=False)
        .head())
```

```python
# ============================================================
# 4. 범주형 변수 통계 검정 결과 요약
# ============================================================

print("\n" + "=" * 60)
print("범주형 변수 통계 검정 결과")
print("=" * 60)

cat_test_sorted = cat_test.sort_values(by='p_value')
display(cat_test_sorted)

print("\n[영향력 기준 상위 변수]")
display(cat_test_sorted.sort_values(by='effect_size_CramersV', ascending=False))
```

```python
# ============================================================
# 5. 숫자형 변수 통계 검정 결과 요약
# ============================================================

print("\n" + "=" * 60)
print("숫자형 변수 통계 검정 결과")
print("=" * 60)

num_test_sorted = num_test.sort_values(by='p_value')
display(num_test_sorted)

print("\n[유의한 숫자형 변수]")
display(num_test_sorted[num_test_sorted['significant_0.05'] == True])
```

```python
# ============================================================
# 6. 변수 세트 검증 결과 요약
# ============================================================

print("\n" + "=" * 60)
print("변수 세트별 교차검증 결과")
print("=" * 60)

feature_set_summary = feature_set_summary.sort_values(by='f1_mean', ascending=False)
display(feature_set_summary)

best_feature_set = feature_set_summary.iloc[0]

print("\n[최종 선택 Feature Set]")
display(best_feature_set.to_frame().T)
```

```python
# ============================================================
# 7. 모델 성능 비교 결과 요약
# ============================================================

print("\n" + "=" * 60)
print("모델 성능 비교 결과")
print("=" * 60)

model_comparison = model_comparison.sort_values(by='f1_mean', ascending=False)
display(model_comparison)

best_model = model_comparison.iloc[0]

print("\n[최고 성능 모델]")
display(best_model.to_frame().T)
```

```python
# ============================================================
# 8. 예측 결과 요약
# ============================================================

print("\n" + "=" * 60)
print("예측 결과 요약")
print("=" * 60)

print("\n[예측 결과 샘플]")
display(submission.head())

print("\n[데이터 형태]")
print(submission.shape)

print("\n[클래스 분포]")
print(submission['Survived'].value_counts())

print("\n[클래스 비율 (%)]")
display((submission['Survived'].value_counts(normalize=True)*100).round(2))
```
